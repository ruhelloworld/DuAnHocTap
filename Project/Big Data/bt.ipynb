- 2.1 Mapreduce (SV phải viết hàm)

```python
from functools import reduce
import re
```

```python
def map(words):
  map = list()
  for i in words:
    map.append(f'<{i}, {1}>')
  return map
```

```python
from os import WCOREDUMP
def reduce(map):
  reduce1 = dict()
  for value in map:
    word = value.strip("<>").split(", ")[0]
    if word not in reduce1.keys():
      reduce1[word] = 1
    else:
      reduce1[word] += 1
  return reduce1
```

 - Trực quan hóa dữ liệu

```python
with open("/content/drive/MyDrive/Colab Notebooks/4/BIG/dataM.txt") as file:
  df = file.read()
df
```

```python
df = df.lower().replace(',', '').replace('.', '').replace(':', '').replace('\n', '')
df = df.split()
```

```python
resultmap = map(df)
for i in resultmap:
  print(i)
```

```python
resultreduce = reduce(resultmap)
for i in resultreduce:
  print(f"<{i}, {resultreduce[i]}>")
```

```python
import matplotlib.pyplot as plt
top_5_words = sorted(resultreduce.items(), key=lambda x: x[1], reverse=True)[:5]

words = [pair[0] for pair in top_5_words]
frequencies = [pair[1] for pair in top_5_words]

# Vẽ biểu đồ
plt.bar(words, frequencies)
plt.xlabel('Từ')
plt.ylabel('Số lần xuất hiện')
plt.title('Top 5 từ xuất hiện nhiều nhất')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
